```python
# Problem 4
# Given list
L = [11, 23, 58, 31, 56, 22, 43, 12, 65, 19]
n = 3

# Step 1: Get all even numbers
evens = []

for num in L:
    if num % 2 == 0:
        evens.append(num)

# Step 2: Remove (n-1) smallest even numbers
for i in range(n - 1):
    minimum = evens[0]

    # Find the minimum in the current evens list
    for value in evens:
        if value < minimum:
            minimum = value

    # Remove the smallest value
    evens.remove(minimum)

# Step 3: Find the new minimum (nth smallest even)
nth_smallest = evens[0]

for value in evens:
    if value < nth_smallest:
        nth_smallest = value

print("The", n, "th smallest even number is:", nth_smallest)
```

I encountered an issue when removing elements from a list while looping over it.

I used remove() inside the same loop that was iterating through the list.
This caused unexpected behavior because the list size changes during iteration.

When elements are removed, the remaining elements shift positions.

This can cause some values to be skipped and lead to logical errors.

Although it worked in a simple case, it is not safe practice.

I fixed the issue by separating the removal process into a controlled loop.
